```python
def plot_correlation_matrix(corr_mat):
    sns.set(style="white")
    mask = np.zeros_like(corr_mat, dtype=bool)
    mask[np.triu_indices_from(mask)] = True
    fig, ax = plt.subplots()
    cmap = sns.diverging_palette(240, 10, n=9, as_cmap=True)
    sns.heatmap(corr_mat, mask=mask, cmap=cmap,
                vmax=.3, center=0, square=True,
                linewidths=.5, cbar_kws={"shrink": .5},
                ax=ax)
    ax.set_title("Correlation Matrix", fontsize=16)
    sns.set(style="ticks")

corr_mat = df.select_dtypes(include="number").corr()
plot_correlation_matrix(corr_mat)
```

```python
#DEcision Tree 
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn import metrics
```

```python
#from Chapter_13_utils import performance_evaluation_report
```

```python
tree_classifier=DecisionTreeClassifier(random_state=42)
tree_classifier.fit(X_train,y_train)
y_pred=tree_classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, roc_curve, auc,accuracy_score,classification_report,precision_recall_curve
```

```python
# Calculate confusion matrix
cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(cm,\
                     index = ['No Default','Default'],\
                     columns = ['No Default','Default'])
```

```python
# Calculate false positive rate, true positive rate, and area under the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(cm_df, annot=True,fmt='g',cmap='RdPu')
plt.title('Decision Tree \nAccuracy:{0:.3f}'\
           .format(accuracy_score(y_test, y_pred)))
plt.ylabel('True Values')
plt.xlabel('Predicted Values')
plt.show()
```

```python
y_pred=tree_classifier.predict(X_test)
target_names=['No Default','Default']
print(classification_report(y_test,y_pred,\
                           target_names=target_names))
```

```python
# ROC curve plot
plt.figure(figsize=(15,8))
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, label='%s AUC = %0.2f' % \
('Decision Tree', roc_auc))
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.ylabel('Sensitivity(True Positive Rate)')
plt.xlabel('1-Specificity(False Positive Rate)')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve
y_pred_prob=tree_classifier.predict_proba(X_test)[:,1]
precision, recall, thresholds = precision_recall_curve(y_test, y_pred_prob)
```

```python
ax = plt.subplot()
ax.plot(recall, precision,
         label=f"PR-AUC = {metrics.auc(recall, precision):.2f}")
ax.set(title="Precision-Recall Curve",
       xlabel="Recall",
       ylabel="Precision")
ax.legend()
```
